```python
from simplet5 import SimpleT5
import pandas as pd
from sacrebleu import corpus_bleu
```

```python
model = SimpleT5()
model.from_pretrained('mt5', 'google/mt5-base')
```

```python
df_ar = pd.read_csv('arawebnlg2020.csv')
df_ar = df_ar.loc[df_ar['category'].isin(['SportsTeam','CelestialBody', 'Food', 'Artist', 'University', 'Politician'])]
df_ar = df_ar.dropna()
df_ar = df_ar[['input_text', 'target_text', 'data']]
print(len(df_ar))
df_ar.head()
```

```python
df_ar_train = df_ar[df_ar['data']=='train']
df_ar_train = df_ar_train[['input_text', 'target_text']]

df_ar_dev = df_ar[df_ar['data']=='dev']
df_ar_dev = df_ar_dev[['input_text', 'target_text']]

df_ar_test = df_ar[df_ar['data']=='test']
df_ar_test = df_ar_test[['input_text', 'target_text']]
```

```python
# simpleT5 expects dataframe to have 2 columns: "source_text" and "target_text"
# T5 model expects a task related prefix: since it is a data-to-text task, we will add a prefix "summarize: "

df_ar_train = df_ar_train.rename(columns={"target_text":"target_text", "input_text":"source_text"})
df_ar_train = df_ar_train[['source_text', 'target_text']]

df_ar_train['source_text'] = "data2text: " + df_ar_train['source_text']
print(df_ar_train)
```

```python
df_ar_dev = df_ar_dev.rename(columns={"target_text":"target_text", "input_text":"source_text"})
df_ar_dev = df_ar_dev[['source_text', 'target_text']]

df_ar_dev['source_text'] = "data2text: " + df_ar_dev['source_text']
print(df_ar_dev)
```

```python
df_ar_test = df_ar_test.rename(columns={"target_text":"target_text", "input_text":"source_text"})
df_ar_test = df_ar_test[['source_text', 'target_text']]

df_ar_test['source_text'] = "data2text: " + df_ar_test['source_text']
print(df_ar_test)
```

```python
model.train(train_df=df_ar_train,
            eval_df = df_ar_dev,
            source_max_token_len = 256,
            target_max_token_len = 256,
            batch_size = 2,
            max_epochs = 5,
            use_gpu = True,
            outputdir = "/output/dir",
            early_stopping_patience_epochs = 3
)
```

```python
input_text = " "
model.predict(input_text)
```
